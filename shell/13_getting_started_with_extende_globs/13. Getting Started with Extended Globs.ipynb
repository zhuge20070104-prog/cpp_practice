```python
# 以Archive 或者Backup开头，以gz,或者xz结尾
cmd='''
shopt -s extglob

ls ./files/@(Archive|Backup)*@(.gz|.xz)

'''
!{cmd}
```

```python
# 像下面这样写也可以，就是长点
cmd='''
shopt -s extglob

ls ./files/Archive*.gz ./files/Archive*.xz ./files/Backup*.gz ./files/Backup*.xz

'''
!{cmd}
```

```python
# 和前面条件一样，但是中间不要有 -2015

cmd='''
shopt -s extglob

ls ./files/@(Archive|Backup)-!(2015)*@(.gz|.xz)

'''
!{cmd}
```

```python
# 2015还是出来了，没用，得改改
# 不能直接*，再加两位，精确一点，再*

cmd='''
shopt -s extglob

ls ./files/@(Archive|Backup)-!(2015)-[0-9][0-9]-*@(.gz|.xz)

'''
!{cmd}
```
