```python
# brace expansion 操作string，不依赖于文件
#  ls with Globs的过程
# bash expands glob for all files in the directory
# ls stats all files
# ls sorts list
# ls displays the list

cmd='''
shopt -s extglob

ls ./files/file* > /dev/null

'''

!{cmd}
```

```python
# brace expansion
cmd='''
shopt -s extglob
ls ./files/file{1..10000} > /dev/null
'''

!{cmd}
```

```python
# echo 不做list，不管文件是否存在
cmd='''
shopt -s extglob
echo file{1..10000} > /dev/null
'''

!{cmd}
```

```python
# 这个glob的match过程
# 先展开成file1*
# file2*
# file3*
# 再展开成 file1.txt
# file2.txt
# file3.txt

cmd='''
shopt -s extglob

ls ./files/file{1..10}*
'''

!{cmd}
```

```python
# ls 和 echo的区别

cmd='''
shopt -s extglob

mkdir files1
touch ./files1/file{1,2,4}.txt

ls ./files1
'''

!{cmd}
```

```python
cmd='''

shopt -s extglob

# 这个ls默认认为有1,2,3,4

ls ./files1/file{1..4}.txt

'''
!{cmd}
```

```python
cmd='''
shopt -s extglob

# echo 这个不报错，他并不确保文件存在

echo files1/files{1..4}.txt

'''

!{cmd}
```

```python
# Brace Expansion的格式

cmd='''

echo {1,2,3}
echo {1..10}
echo {,.bak}
'''

!{cmd}
```

```python
# 这种符号也可以组合
cmd='''
shopt -s extglob

echo {a..c}{10..39}
'''
!{cmd}
```

```python
#  这样组合亦可
cmd='''
shopt -s extglob

echo {a,b,c}{d,e,f}
'''

!{cmd}
```

```python
# 1-100, 以2为间隔

cmd='''

shopt -s extglob

echo {1..100..2}
'''

!{cmd}
```

```python
# 相当于np.linspace
cmd='''

shopt -s extglob

echo {2..100..2}
'''

!{cmd}
```

```python
# 也可以在左边pad 0

cmd='''

shopt -s extglob

echo {0002..1000..2}
'''

!{cmd}
```

```python
# 字符也可以


cmd='''

shopt -s extglob

echo {a..z}
'''

!{cmd}
```

```python
#  Skip letter也可以

cmd='''

shopt -s extglob

echo {a..z..2}
'''

!{cmd}
```

```python
#  倒过来也行，10-0
cmd='''

shopt -s extglob

echo {10..0}
'''

!{cmd}
```

```python
# 可以匹配file,也可以匹配file.bak

cmd='''
echo ./files/file{,.bak}
'''

!{cmd}
```

```python
# 这个echo 不检查存在与否

cmd='''
echo /this/is/a/long/path/file{,.bak}
'''
!{cmd}
```
