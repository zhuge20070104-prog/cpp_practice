```python
import numpy as np
import pandas as pd
import matplotlib
%matplotlib inline
```

```python
d = pd.date_range('1/2/2020', periods=500)
d
```

```python
ts = pd.Series(np.random.randn(500), index=d)
ts
```

```python
# cumsum,是求到目前为止的各项累加和
cumsum_ts = ts.cumsum()
cumsum_ts
```

```python
# series 可以直接plot
cumsum_ts.plot()
```
