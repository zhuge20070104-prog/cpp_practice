```python
import numpy as np
import pandas as pd
```

```python
d = pd.date_range('20200301', periods=10)
df = pd.DataFrame(np.random.randn(10, 4), index=d, columns=['A', 'B', 'C', 'D'])
df
```

```python
# 构造含有Nan值的DataFrame
df2 = df.reindex(index=d[0:4], columns=list(df.columns) + ['E'])
df2.loc[d[0]:d[1], 'E'] = 1.0
df2
```

```python
# 检查null值
df2.isnull()
```

```python
# 删除null 值
df2.dropna()
```

```python
# 用其他值填充Nan
df2.fillna(value=2.0)
```

```python
# 另外一种check null值的方法
pd.isna(df2)
```
