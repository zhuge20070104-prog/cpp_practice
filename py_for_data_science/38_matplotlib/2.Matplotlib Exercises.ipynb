## Data

```python
import numpy as np
x = np.arange(0, 100)
y = x * 2
z = x ** 2
```

```python
# import
import matplotlib.pyplot as plt
%matplotlib inline
```

## Exercise3

** Create the plot below by adding two axes to a figure object at [0,0,1,1] and [0.2,0.5,.4,.4]**

```python
fig = plt.figure()
fig.add_axes([0, 0, 1, 1])
fig.add_axes([0.2, 0.5, 0.4, 0.4])
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0, 0, 1, 1])
ax1.plot(x, z)
ax1.set_xlabel('x')
ax1.set_ylabel('z')

ax2 = fig.add_axes([0.2, 0.5, 0.4, 0.4])
ax2.plot(x, y)
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_title('zoom')
ax2.set_xlim(20, 22)
ax2.set_ylim(30, 50)
```

## Exercise 4

** Use plt.subplots(nrows=1, ncols=2) to create the plot below.**

```python
fig, ax = plt.subplots(nrows=1, ncols=2)
plt.tight_layout()
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2)
ax[0].plot(x, y, 'b--', lw=3)
ax[1].plot(x, z, 'r-', lw=3)
plt.tight_layout()    
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(12, 3))
ax[0].plot(x, y, 'b-', lw=3)
ax[0].set_xlabel('x')
ax[0].set_ylabel('y')
ax[1].plot(x, z, 'r--', lw=3)
ax[1].set_xlabel('x')
ax[1].set_ylabel('z')
plt.tight_layout()    
```
