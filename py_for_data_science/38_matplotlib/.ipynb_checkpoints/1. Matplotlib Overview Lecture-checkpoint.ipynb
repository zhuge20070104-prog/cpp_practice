```python
import matplotlib.pyplot as plt
%matplotlib inline
```

## Basic Example

```python
# 我们想要绘制的数据
import numpy as np

x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
x
```

```python
y
```

## Basic Matplotlib Commands

```python
plt.xlabel('X Axis Title Here')
plt.ylabel('Y Axis Title Here')
plt.title('String Title Here')
plt.plot(x, y, 'r') # 'r' is the color red
```

##  同一画布画多张图


```python
# plt subplot 参数: nrows, ncols, plot_number
plt.subplot(1, 2, 1)
# 红色，点划线
plt.plot(x, y, 'r--')

plt.subplot(1, 2, 2)
# 绿色, *点， 实线
plt.plot(y, x, 'g*-')
```

## Matplotlib 面向对象模式

### 引入面向对象模式

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])

axes.set_xlabel('Set X Label')
axes.set_ylabel('Set Y Label')
axes.set_title('Set Title')
axes.plot(x, y, 'b')
```

```python
# 代码有点复杂， 但是好处是，现在我们可以完全控制plot的axes放置在哪儿，我们可以在图上放置多于一个的axes

fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])  # [left, bottom, width, height] 
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])  

# 大的图 axes1
axes1.plot(x, y, 'b')
axes1.set_xlabel('X_label axes1')
axes1.set_ylabel('Y_label axes1')
axes1.set_title('Axes 1 Title')


# 小的图 axes2
axes2.plot(y, x, 'r')
axes2.set_xlabel('X_label axes2')
axes2.set_ylabel('Y_label axes2')
axes2.set_title('Axes 2 Title')
```

## 面相对象绘图使用subplots

```python
fig, axes = plt.subplots()
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title')
axes.plot(x, y, 'r')
```

```python
# Subplots的时候可以指定行数和列数
fig, axes = plt.subplots(nrows=3, ncols=3)
plt.tight_layout()
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
```

```python
# axes 是一个用于绘制的axes数组
axes
```

```python
# 我们可以迭代这个数组
for ax in axes:
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('title')
    ax.plot(x, y, '-r')

# 显示figure对象
fig
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x, y)
axes[0].set_title('Has title')
axes[1].plot(y, x)
```

```python
# 使用matplotlib的一个通常的问题就是图像的overlapping，我们可以使用plt.tightlayout来避免图像overlapping
fig, axes = plt.subplots(nrows=1, ncols=2)

for ax in axes:
    ax.plot(x, y, 'g')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('title')

fig
plt.tight_layout()
```

## Figure size, aspect ratio and DPI

```python
fig = plt.figure(figsize=(8, 4), dpi=100)
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y , '-g')
```

```python
# 同样的参数也可以传递给layout 管理器，像subplots
fig, axes = plt.subplots(figsize=(12, 3))

axes.plot(x, y, 'r')

axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('Sample plot')
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(6, 4))
for ax in axes:
    ax.set_title('Sample plot')
    ax.plot(x, y)

plt.tight_layout()
```

## Legends

```python
# 先用label参数加上legends，最后ax.legend() 一下就可以了
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, x**2, label='x^2')
ax.plot(x, x**3, label='x^3')
# legend的位置也可以设置，使用loc参数就可以
ax.legend()
```

```python
# legend location的含义： https://matplotlib.org/2.0.2/api/pyplot_api.html#matplotlib.pyplot.legend

ax.legend(loc=1) # 右上角
fig

ax.legend(loc=2) # 左上角
fig

ax.legend(loc=3) # 左下角
fig

ax.legend(loc=4) # 右下角
fig

ax.legend(loc=0) # best，让matplotlib自己确定哪个最好
fig
```

## Setting Colors, LineWidths, LineTypes

```python
# Matlab 样式的color and style
fig, ax = plt.subplots()
ax.plot(x, x**2, 'b.-') # 蓝色点线
ax.plot(x, x**3, 'g--') # 绿色虚线
```

### 使用color参数指定color

```python
fig, ax = plt.subplots()
ax.plot(x, x+1, color='blue', alpha=0.5) # 半透明
ax.plot(x, x+2, color='#8B008B')
ax.plot(x, x+3, color='#FF8C00')
```

### Line and Marker Styles

```python
fig, ax = plt.subplots(figsize=(12, 6))

# Linewidth
ax.plot(x, x+1, color='red', linewidth=0.25)
ax.plot(x, x+2, color='red', linewidth=0.50)
ax.plot(x, x+3, color='red', linewidth=1.00)
ax.plot(x, x+4, color='red', linewidth=2.00)


# 可能的linestyle选项， ‘-‘, ‘–’, ‘-.’, ‘:’, ‘steps’
ax.plot(x, x+5, color='green', lw=3, linestyle='-')
ax.plot(x, x+6, color='green', lw=3, ls='-.')
ax.plot(x, x+7, color='green', lw=3, ls=':')

# 自定义dash
line, = ax.plot(x, x+8, color='black', lw=1.50)
line.set_dashes([5, 10, 15, 10]) # format, line length, space length ...

# 可能的marker 选项, marker = '+', 'o', '*', 's', ',', '.', '1', '2', '3', '4', ...
ax.plot(x, x+9, color='blue', lw=3, ls='-', marker='+')
ax.plot(x, x+10, color='blue', lw=3, ls='--', marker='o')
ax.plot(x, x+11, color='blue', lw=3, ls='-', marker='s') # square 正方形
ax.plot(x, x+12, color='blue', lw=3, ls='--', marker='1')


# marker size and color
ax.plot(x, x+13, color='purple', lw=1, ls='-', marker='o', markersize=2)
ax.plot(x, x+14, color='purple', lw=1, ls='-', marker='o', markersize=4)
ax.plot(x, x+15, color='purple', lw=1, ls='-', marker='o', markersize=8, markerfacecolor='red')
ax.plot(x, x+16, color='purple', lw=1, ls='-', marker='s', markersize=8,
       markerfacecolor='yellow', markeredgewidth=3, markeredgecolor='green')
```

## Control over axes appearance

```python
fig, axes = plt.subplots(1, 3, figsize=(12, 4))
axes[0].set_title('Default axes range')
axes[0].plot(x, x**2, x, x**3)

axes[1].axis('tight')
axes[1].set_title('tight axes')
axes[1].plot(x, x**2, x, x**3)

# Set axis limit for this plot
axes[2].set_ylim([0, 60])
axes[2].set_xlim([2, 5])
axes[2].set_title('custom axes range')
axes[2].plot(x, x**2, x, x**3)
```
